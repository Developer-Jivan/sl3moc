```python
#importing necessary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Reading dataset

df=pd.read_csv("3ass_student_data.csv")

print("Dataset Preview")

print(df.head())
```

```python
print("\nOverall Statistics:\n")

print(df.describe())
```

```python
#checking for missing values
print("\nMissing Values:\n")

print(df.isnull().sum())
```

```python
+#correlation heatmap
newdf=df.select_dtypes(include=['number'])

plt.figure(figsize=(8,5))

sns.heatmap(newdf.corr(),annot=True,cmap='coolwarm')

plt.show() 
```

```python
#Mean grouped by Gender
print("\nMean Marks grouped by Gender:\n")

print(df.groupby("Gender")["Marks"].mean())
```

```python
#Median grouped by Department
print("\nMedian Marks grouped by Department:\n")

print(df.groupby("Department")["Marks"].median())
```

```python
#Standard deviation grouped by Gender and Department
#Standard deviation shows how marks vary in each group.
print("\nStandard deviation grouped by Gender and Department:\n")

print(df.groupby(["Gender","Department"])["Marks"].std())
#It calculates how much marks vary (spread) for each group of Gender and Department
```

```python
#Mean grouped by Result
print("\nMean Marks grouped by Result:\n")

print(df.groupby("Result")["Marks"].mean())
```

```python
#Mean grouped by Result and Gender
print("\nMean Marks grouped by Result and Gender:\n")

print(df.groupby(["Result","Gender"])["Marks"].mean())
```

```python
#Count categorical values 
print("\nCount of Departments:\n")

print(df["Department"].value_counts())
```

```python
#minimum marks by gender 
print("\nMinimum Marks by Gender:\n")

print(df.groupby("Gender")["Marks"].min())
```

```python
#maximum marks by gender 
print("\nMaximum Marks by Gender:\n")

print(df.groupby("Gender")["Marks"].max())
```

```python
# list of marks
male_marks = df[df["Gender"] == "Male"]["Marks"].tolist()
female_marks = df[df["Gender"] == "Female"]["Marks"].tolist()

print("Male Marks:", male_marks)
print("Female Marks:", female_marks)
```

```python
# list of marks by department
dept_marks = df.groupby("Department")["Marks"].apply(list)
print("\nMarks by Department:\n", dept_marks)
```

# ======= Iris Dataset Descriptive Statistics =======

```python
# Load the dataset
iris_df = pd.read_csv("3ass_iris.csv")

print("Iris dataset preview:\n")
print(iris_df.head())
```

```python
# Overall Dataset Statistics
print("Overall Dataset Statistics:")
print(iris_df.describe())
```

```python
print("\nMissing Values:\n")
print(iris_df.isnull().sum())
```

```python
#Check Unique Species
print("Unique species:")
print(iris_df['species'].unique())
```

```python
# Grouped summary statistics
grouped = iris_df.groupby('species')

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)    

print("summary statistics for each  species:")
print(grouped.describe())
```

```python
# Statistical Measures for Each Species
for species in ['Iris-setosa','Iris-versicolor','Iris-virginica']:
    
    if species not in grouped.groups:
        print("Species not found")
        continue
    
    sp = grouped.get_group(species)
    
    print("\n=====", species, "=====")
    
    print("Mean:")
    print(sp.mean(numeric_only=True))
    
    print("\nStandard Deviation:")
    print(sp.std(numeric_only=True))
    
    print("\n25th Percentile:")
    print(sp.quantile(0.25, numeric_only=True))   
    
    print("\nMedian (50th Percentile):")
    print(sp.quantile(0.5, numeric_only=True))    
    
    print("\n75th Percentile:")
    print(sp.quantile(0.75, numeric_only=True))
```

```python
# Overall Percentiles of Dataset
print("\nOverall Percentiles of Dataset:")
print(iris_df.quantile([0.25, 0.5, 0.75], numeric_only=True))
```
