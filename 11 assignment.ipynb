Create databases and tables, insert small amounts of data, and run simple queries using Impala

```python
# Importing Packages
import sqlite3
import pandas as pd
```

```python
# Connecting to the Impla Server
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print("Database created successfully!")
```

```python
cursor.execute("""
CREATE TABLE students (
    student_id INTEGER,
    name TEXT,
    age INTEGER,
    course TEXT,
    marks REAL
)
""")

print("Table 'students' created!")
```

```python
students_data = [
    (1, 'Rohan', 21, 'AI', 85.5),
    (2, 'Amit', 22, 'DS', 78.0),
    (3, 'Priya', 20, 'AI', 92.3),
    (4, 'Sneha', 21, 'ML', 88.8),
    (5, 'Karan', 23, 'DS', 67.5)
]

cursor.executemany("INSERT INTO students VALUES (?, ?, ?, ?, ?)", students_data)

conn.commit()
print("Data inserted successfully!")
```

```python
df = pd.read_sql_query("SELECT * FROM students", conn)
df
```

```python
df = pd.read_sql_query("SELECT name, course FROM students", conn)
df
```

```python
df = pd.read_sql_query("SELECT * FROM students WHERE marks > 80", conn)
df
```

```python
df = pd.read_sql_query("SELECT AVG(marks) AS average_marks FROM students", conn)
df
```

```python
cursor.execute("""
UPDATE students
SET marks = 70
WHERE student_id = 5
""")

conn.commit()

df = pd.read_sql_query("SELECT * FROM students", conn)
df
```

```python
cursor.execute("DELETE FROM students WHERE student_id = 2")

conn.commit()

df = pd.read_sql_query("SELECT * FROM students", conn)
df
```

```python
cursor.execute("DROP TABLE students")

print("Table dropped successfully!")
```

```python
conn.close()
print("Connection closed!")
```
